```python
import pandas as pd
import numpy as np
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

```python
snails = pd.DataFrame({"age":np.random.normal(10,1,size=(100))})
snails['weight'] = ( 
    snails['age'] * 3 # proportional to age
    + 5               # with an offset - they start weighing 5
    + np.random.normal(0,1,size=(100)) ) #and they don't all grow the same way - add some noise
```

```python
snails['weight_noisy'] = (snails['age'] * 3 + 5 + np.random.normal(0,20,size=(100)) )
snails['weight_clean'] = (snails['age'] * 3 + 5 + np.random.normal(0,0,size=(100)) )
```

```python
sns.lmplot(x='age',y='weight',data=snails,fit_reg=False)
```

```python
sns.lmplot(x='age',y='weight',data=snails)
```

```python
sns.lmplot(x='age',y='weight',data=snails.loc[0:20])
```

```python
snails['weight_noisy'] = (snails['age'] * 3 + 5 + np.random.normal(0,20,size=(100)) )
snails['weight_clean'] = (snails['age'] * 3 + 5 + np.random.normal(0,0,size=(100)) )
```

```python
sns.lmplot(x='age',y='weight_noisy',data=snails)
```

```python
sns.lmplot(x='age',y='weight_clean',data=snails)
```

```python

fig, axs = plt.subplots(ncols=4,figsize=(18, 5))
sns.regplot(x='age', y='weight', data=snails, ax=axs[0])
sns.regplot(x='age', y='weight_noisy', data=snails, ax=axs[1])
sns.regplot(x='age',y='weight_clean', data=snails, ax=axs[2])
sns.regplot(x='age',y='weight', data=snails.loc[0:20], ax=axs[3])
```

```python
stats.linregress(snails['age'],snails['weight'])
```

```python
stats.linregress(snails['age'],snails['weight_clean'])
```

```python
stats.linregress(snails['age'],snails['weight_noisy'])
```

```python
stats.linregress(snails['age'][0:20],snails['weight'][0:20])
```

```python
fig, axs = plt.subplots(ncols=4,figsize=(18, 5))
sns.residplot(x='age', y='weight', data=snails, ax=axs[0])
sns.residplot(x='age', y='weight_noisy', data=snails, ax=axs[1])
sns.residplot(x='age',y='weight_clean', data=snails, ax=axs[2])
sns.residplot(x='age',y='weight', data=snails.loc[0:20], ax=axs[3])
```
